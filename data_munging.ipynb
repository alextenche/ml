### dealing with big datasets

```python
iris_chunks = pd.read_csv('data/iris.csv', names=['C1', 'C2', 'C3', 'C4', 'C5'], chunksize=10)
for chunk in iris_chunks:
    print('Shape: ', chunk.shape)
    print(chunk, '\n')
```

```python
iris_iterator = pd.read_csv('data/iris.csv', header=0, names=['C1', 'C2', 'C3', 'C4', 'C5'], iterator=True)
```

```python
iris_iterator.get_chunk(10)
```

```python
iris_iterator.get_chunk(20).shape
```

```python
import csv
with open('data/iris.csv', 'rt') as data_stream:
    for n, row in enumerate(csv.DictReader(data_stream, fieldnames=['sepal_length','sepal_width','petal_length','petal_width'],dialect='excel')):
        if n == 0:
            print(n, row)
        else:
            break
```
